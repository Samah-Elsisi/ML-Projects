# Seaborn : violinplot

```python
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
cars = sns.load_dataset('mpg').dropna()

cars.shape
```

```python
cars.head()
```

```python
sns.set_style('whitegrid')
```

```python
cars.cylinders.value_counts()
```

```python
cars = cars[cars.cylinders.isin([4,6,8])]
```

```python
sns.violinplot(cars.displacement)
```

```python
sns.violinplot(x=cars.cylinders , y=cars.displacement)
```

```python
sns.violinplot(x='cylinders' , y='displacement' , data=cars)
```

```python
sns.violinplot(x='cylinders' , y='displacement' , data=cars , hue='origin')
```

## Violinplot Options

```python
sns.violinplot(x='cylinders' , y='displacement' , hue='origin',
              data=cars[cars.origin.isin(['japan' , 'europe'])]
              )
```

### split

```python
sns.violinplot(x='cylinders' , y='displacement' , hue='origin' , 
              data=cars[cars.origin.isin(['japan' , 'europe'])] ,
              split=True
              )

plt.legend(loc=2)
```

### inner

```python
sns.violinplot(x='cylinders' , y='displacement' , hue='origin' ,
              data=cars[cars.origin.isin(['japan' , 'europe'])] ,
              split=True ,
              inner ='quartiles'
              )

plt.legend(loc=2)
```

### scale

```python
sns.violinplot(x='cylinders' , y='displacement' , hue='origin' ,
              data=cars[cars.origin.isin(['japan' , 'europe'])] ,
              split=True ,
              scale ='count'
              )

plt.legend(loc=2)
```

```python
sns.violinplot(x='cylinders' , y='displacement' , hue='origin' ,
              data=cars[cars.origin.isin(['japan' , 'europe'])] ,
              split=True ,
              scale ='count',
              scale_hue=False
              )

plt.legend(loc=2)
```

```python
(
    cars[cars.origin.isin(['japan' , 'europe'])]
    .groupby('cylinders')
    .origin
    .value_counts()
)
```

## Styling

### Box Options

```python
sns.violinplot(x=cars.origin , y=cars.displacement ,
              order=['japan' , 'europe' , 'usa']
              )
```

```python
sns.violinplot(x=cars.origin , y=cars.displacement ,
              linewidth=5
              )
```

```python
sns.violinplot(x=cars.origin , y=cars.displacement ,
              linewidth=0.2
              )
```

# Seaborn: scatterplot

```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
diamonds =sns.load_dataset('diamonds')

diamonds.shape
```

```python
diamonds.head()
```

```python
diamonds = diamonds[
    diamonds.cut.isin(['Premium' , 'Good']) &
    diamonds.color.isin(['D' ,'F' ,'J'])
    ].sample(n=100)

diamonds.shape
```

## Basics

```python
sns.set_style('dark')
```

```python
sns.scatterplot(diamonds.carat , diamonds.price)
```

```python
sns.scatterplot(x='carat' , y='price', data=diamonds)
```

```python
sns.scatterplot(x='x' , y='y' , data=diamonds)
```

## Semantic Variables

```python
sns.set_style('darkgrid')
```

### hue

```python
sns.scatterplot(x='carat' , y='price' , hue='cut' ,
               data=diamonds)
```

```python
sns.scatterplot(x='carat', y='price', hue='cut',
                data=diamonds,
                palette = ['purple', '#55CCCC'],
                hue_order=['Good', 'Premium']
               )
```

## Continouns Variables

```python
sns.scatterplot(x='carat' , y='price' , hue='depth',
               data=diamonds
               )
```

### style

```python
sns.scatterplot(x='carat' , y='price' , style='cut' ,
               data=diamonds ,
                s=100
               )
```

### size

```python
sns.scatterplot(x='carat' , y='price' , size='cut' ,
               data=diamonds
               )
```

```python
sns.scatterplot(x='carat' , y='price' , size='depth' ,
               data=diamonds
               )
```

### hue + style + size

```python
sns.scatterplot(x='carat' , y='price' ,
               hue='cut' ,
               style='cut' ,
               data=diamonds,
               s=100
               )
```

```python
sns.scatterplot(x='carat' , y='price' ,
               hue='cut' ,
               style='color' ,
               data=diamonds,
               s=100
               )
```

```python
sns.scatterplot(x='carat' , y='price' ,
               hue='cut' ,
               style='color' ,
               data=diamonds,
               size='depth'
               )
```

## Additional Styling

### alpha

```python
sns.scatterplot(x='carat' , y='price' ,
              data=diamonds ,
              s=100 ,
              alpha=0.6
              )
```
